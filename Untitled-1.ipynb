```python
import matplotlib.pyplot as plt

# Datos
dias_semana = ['Lunes', 'Martes', 'Miércoles',
               'Jueves', 'Viernes', 'Sábado', 'Domingo']
hamburguesas_comidas = [3, 5, 2, 4, 7, 6, 2]

# Crear la gráfica de barras
plt.bar(dias_semana, hamburguesas_comidas, color='blue')

# Títulos y etiquetas
plt.title('Hamburguesas Comidas por Día de la Semana')
plt.xlabel('Día de la Semana')
plt.ylabel('Cantidad de Hamburguesas')

# Mostrar la gráfica
plt.show()

print('This')

x = 20
x = 25
x = 23
```
